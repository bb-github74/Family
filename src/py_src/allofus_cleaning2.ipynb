### Patient with Some Types of Liver Cancer

```python
hasLiverCancer = pd.read_csv('D:F2023/AOU/df_B.csv')
```

```python
print('Number of Conditions')
hasLiverCancer.standard_concept_name.value_counts()
```

```python
def get_earliest_date_lcancer(dataframe: pd.DataFrame= None):
    '''
    params: 
    accepts dataframe and get earliest date based on person id.
    The person with more than 2 conditions, will get the only earliest date 
    of the condition. We are not interested in what is the name of the condition.

    returns: dataframe consists of person id and earliest date
    '''
    output = {'person_id':[], 'csdLC':[]}
    dataframe['csd'] = dataframe.condition_start_datetime.apply(lambda x: pd.to_datetime(x, utc=True))
    ids = dataframe.person_id.unique()
    for id in ids:
        df = dataframe[dataframe.person_id == id]
        output['person_id'].append(id)
        output['csdLC'].append(df.csd.min())
    return output
```

```python
hasLC = pd.DataFrame(get_earliest_date_lcancer(dataframe=hasLiverCancer))
```

```python
hasLC.columns = ['pid', 'csdLC']
hasLC['value'] = 1
hasLC
```

```python
df = pd.read_csv('D:/F2023/AOU/df_C.csv')
```

```python
df.columns = ['pid', 'scn', 'scc', 'csd']
df
```

```python
def simpler_concept(x: str=None)-> str:
    val = ''
    if x == 'Disorder of musculoskeletal system':
        val = 'musculoskeletal'
    elif x == 'Disorder of digestive system':
        val = 'digestive'
    elif x == 'Disorder of breast':
        val = 'breast'
    elif x == 'Disorder of hematopoietic structure':
        val = 'hematopoietic'
    elif x == 'Disorder of endocrine system':
        val = 'endocrine'
    elif x == 'Disorder of cardiovascular system': 
        val = 'cardiovascular'
    elif x == 'Disorder of respiratory system':
        val = 'respiratory'
    elif x == 'Disorder of integument':
        val = 'integument'
    elif x == 'Disorder of nervous system':
        val = 'nervous'
    elif x == 'Disorder of the genitourinary system':
        val = 'genitourinary'
    elif x == 'Disorder of mediastinum':
        val = 'mediastinum'
    elif x == 'Disorder of immune structure':
        val = 'immune'
    elif x == 'Disorder of mucous membrane':
        val = 'mucous'
    else: 
        val = 'disease not in'
        
    return val
```

```python
def get_earliest_date(dataframe: pd.DataFrame= None)-> pd.DataFrame:
    '''
    params: 
    accepts dataframe and get earliest date based on person id.
    The person with more than 2 conditions, will get the only earliest date 
    of the condition. We are not interested in what is the name of the condition.

    returns: dataframe consists of person id and earliest date
    '''
    output = []
    ids = dataframe.pid.unique()
    for ind, id in enumerate(ids):
        #print(ind, id)
        
        df = dataframe[dataframe.pid == id] 
        df_sorted = df.sort_values(by='csd')
        if df_sorted.shape[0] == 1:
            output.append(df_sorted[['pid', 'scn','csd']])
        else:
            output.append(df_sorted[['pid', 'scn','csd']].head(1))
    return output
```

```python
# list to data frame using pd.concat across rows
df_earliest = pd.concat(get_earliest_date(dataframe=df), axis=0)
df_earliest
```

```python
df_earliest = df_earliest.reset_index(drop=True)
df_earliest
```

```python
def istype_csdLC(x):
    if isinstance(x, pd._libs.tslibs.nattype.NaTType):
        return 0
    else:
        return 1
```

```python
# 
combined['csdLC_type'] = combined.csdLC.apply(lambda x: istype_csdLC(x))
```

```python
# comparing two cell values must be the same: datetime or pd.to_datetime
def earlier(dataframe: pd.DataFrame = None) -> list:
    '''

    '''
    output = []
    for row in dataframe.to_dict('records'):
        if isinstance(row['csdLC'], pd._libs.tslibs.nattype.NaTType):
            # some are empty non timestamp values
            output.append('novalue')
        else:
            if row['csd'] < row['csdLC']:
                output.append('before')
            else:
                output.append('after')
    
    return output
```

```python
# attach the values back to dataframe
combined['earlier'] = earlier(dataframe=combined)
```
